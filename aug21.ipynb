```python
import os 
rootdir = 'D:\\Dropbox\\Dropbox\\2016\\Methylation\\Aggregate\\m1A\\'

a = []
c = []
g = []
t = []

for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        if 'PlusOneA' in file:
            aname = os.path.join(subdir, file)
            afile = pd.read_csv(aname, sep = '\t', header = None)
            a = afile[0]
        elif 'PlusOneC' in file:
            cname = os.path.join(subdir, file)
            cfile = pd.read_csv(cname, sep = '\t', header = None)
            c = cfile[0]
        elif 'PlusOneG' in file:
            gname = os.path.join(subdir, file)
            gfile = pd.read_csv(gname, sep = '\t', header = None)
            g = gfile[0]
        elif 'PlusOneT' in file:
            tname = os.path.join(subdir, file)
            tfile = pd.read_csv(tname, sep = '\t', header = None)
            t = tfile[0]
            
weights = [a,c,g,t]
plt.boxplot(weights)
plt.title('Mutational Weight Aggregate on m1A +1 Nucleotide')
plt.xticks([1,2,3,4],['+1 A','+1 C','+1 G', '+1 T'])
plt.ylim(0,1)

plt.savefig('m1APlusOne.png', bbox_inches='tight')
```
